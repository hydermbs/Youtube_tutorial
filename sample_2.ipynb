```python
import requests
from bs4 import BeautifulSoup as bs 
import pandas as pd
```

```python
url = 'https://webscraper.io/test-sites/e-commerce/allinone/computers/laptops'
response = requests.get(url)
print(response)
```

```python
page = bs(response.text,'html.parser')
print(page)
```

```python
big_box = page.find_all('div',{'class':'col-md-4 col-xl-4 col-lg-4'})
print(big_box)
```

```python
data_list = []
for box in big_box:
    name = box.find('a',{'class':'title'}).text
    print(name)
    price = box.find('h4',{'class':'price float-end card-title pull-right'}).text
    desc = box.find('p',{'class':'description card-text'}).text
    reviews = box.find('p',{'class':'review-count float-end'}).text
    print(price)
    print(desc)
    print(reviews)
    dict_data = {'Name':name,
                 'Price':price,
                 'Desc':desc,
                 'Reviews':reviews}
    data_list.append(dict_data)
```

```python
df = pd.DataFrame(data_list)
df.to_csv('Sample_data.csv')
```
